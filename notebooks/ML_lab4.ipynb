### Реализация персептрона

```python
class Activation:
    def callFunction(x):
        pass
    def callDerivative(x):
        pass

class Relu(Activation):
    @staticmethod
    def callFunction(x):
        return np.maximum(0,x)
    @staticmethod
    def callDerivative(x):
        return np.where(x > 0, 1.0, 0.0)

class Sigmoid(Activation):
    @staticmethod
    def callFunction(x):
        return 1 / (1 + np.exp(-x))
    @staticmethod
    def callDerivative(x):
        return Sigmoid.callFunction(x) * (1- Sigmoid.callFunction(x))
class TangH(Activation):
    
    @staticmethod
    def callFunction( x):
        return (np.exp(x) - np.exp(-x)) / (np.exp(x)+np.exp(-x))
    
    @staticmethod
    def callDerivative(x):
        return 1 - TangH.callFunction(x) ** 2

```

```python
class Layer:
    def __init__(self):
        self.input = None
        self.output = None
    
    def forward_propagation(self, input):
        raise NotImplementedError

    def backward_propagation(self, output_error):
        raise NotImplementedError
    
class FCLayer(Layer):

    def __init__(self, input_size, output_size):
        self.weights = np.random.rand(input_size, output_size) - 0.5
        self.bias = np.random.rand(1, output_size) - 0.5

    def forward_propagation(self, input_data):
        self.input = input_data
        self.output = np.dot(self.input, self.weights) + self.bias
        return self.output

    def backward_propagation(self, output_error, learning_rate):
        input_error = np.dot(output_error, self.weights.T)#dE/dx
        weights_error = np.dot(self.input.T, output_error)#dE/dw
        self.weights -= learning_rate * weights_error
        self.bias -= learning_rate * output_error
        return input_error #dE/dB так как равен de/dx

class ActivationLayer(Layer):
    def __init__(self, activation, activation_prime):
        self.activation = activation
        self.activation_prime = activation_prime

    def forward_propagation(self, input_data):
        self.input = input_data
        self.output = self.activation(self.input)
        return self.output

    def backward_propagation(self, output_error, learning_rate):
        return self.activation_prime(self.input) * output_error
```

```python
class LossFunction:
    def fun():
        pass
    def derivative():
        pass

class MeanSquaredError(LossFunction):
    def fun(y_true, y_pred):
        return np.mean(np.power(y_true-y_pred, 2))
    
    def derivative(y_true, y_pred):
        return 2*(y_pred-y_true)/y_true.size
```

```python
class NN:
    def __init__(self):
        self.layers = []
        self.loss = None
        self.loss_prime = None

    def add(self, layer):
        self.layers.append(layer)

    def compile(self, loss, loss_prime):
        self.loss = loss
        self.loss_prime = loss_prime

    def predict(self, input_data):
        samples = len(input_data)
        result = []

        for i in range(samples):
            output = input_data[i]
            for layer in self.layers:
                output = layer.forward_propagation(output)
            result.append(*output)

        return result

    def fit(self, x_train, y_train, epochs, learning_rate):

        samples = len(x_train)

        for i in range(epochs):

            for j in range(samples):
                output = x_train[j]
                for layer in self.layers:
                    output = layer.forward_propagation(output)

                error = self.loss_prime(y_train[j], output)
                for layer in reversed(self.layers):
                    error = layer.backward_propagation(error, learning_rate)
                    
            
```

### Классификация

```python
df = pd.read_csv("../data/classification/csgo_task_update.csv", index_col=0).drop_duplicates()
df
```

```python
X_clf = df.drop(["bomb_planted"], axis=1).values
y_clf = df["bomb_planted"].values
```

```python
std = StandardScaler()
std.fit(X_clf)
X_clf = pd.DataFrame(std.transform(X_clf), columns=df.drop(["bomb_planted"], axis=1).columns).values
```

```python
underSampler = RandomUnderSampler(random_state=42)
X_clf, y_clf = underSampler.fit_resample(X_clf,y_clf)
X_clf= X_clf.reshape(X_clf.shape[0], 1, X_clf.shape[1])
```

```python
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(X_clf,
                                                                        y_clf,
                                                                        stratify=y_clf,
                                                                        test_size=0.2)
```

```python
model_classification = NN()
model_classification.add(FCLayer(17, 3))
model_classification.add(ActivationLayer(TangH.callFunction, TangH.callDerivative))
model_classification.add(FCLayer(3, 1))
model_classification.add(ActivationLayer(Sigmoid.callFunction, Sigmoid.callDerivative))
model_classification.compile(MeanSquaredError.fun, MeanSquaredError.derivative)
model_classification.fit(X_clf_train, y_clf_train, epochs=50, learning_rate=0.001)
pred_classification = model_classification.predict(X_clf_test)
```

```python
pred_classification
```

```python
print(classification_report(y_clf_test, np.around(pred_classification)))
print(confusion_matrix(y_clf_test, np.around(pred_classification)))
```

```python
df_reg = pd.read_csv("../data/regression/trip_duration_update.csv", index_col=0).drop_duplicates()
df_reg
```

```python
y_reg = df_reg["trip_duration"].values
X_reg = df_reg.drop(["trip_duration"], axis=1).values.reshape(df_reg.shape[0], 1, df_reg.shape[1] - 1)
```

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2,random_state=4)
X_train_reg.shape, X_test_reg.shape, y_train_reg.shape, y_test_reg.shape 
```

```python
model_reg = NN()
model_reg.add(FCLayer(4, 16))
model_reg.add(ActivationLayer(Relu.callFunction, Relu.callFunction))
model_reg.add(FCLayer(16, 8))
model_reg.add(ActivationLayer(Relu.callFunction, Relu.callFunction))
model_reg.add(FCLayer(8, 4))
model_reg.add(ActivationLayer(Relu.callFunction, Relu.callFunction))
model_reg.add(FCLayer(4, 1))
model_reg.compile(MeanSquaredError.fun, MeanSquaredError.derivative)
model_reg.fit(X_train_reg, y_train_reg, epochs=50, learning_rate=0.001)
```

```python
pred_reg = model_reg.predict(X_test_reg)
```

```python
print(mean_absolute_error(y_test_reg, pred_reg))
print(mean_squared_error(y_test_reg, pred_reg))
print(r2_score(y_test_reg, pred_reg))
print(max_error(y_test_reg, pred_reg))
```
